### Create the Contacts DataFrame 
---
**Create a Contacts DataFrame that has the following columns:**
- A column named "contact_id"  that contains the unique number of the contact person.
- A column named "first_name" that contains the first name of the contact person.
- A column named "last_name" that contains the first name of the contact person.
- A column named "email" that contains the email address of the contact person

Then export the DataFrame as a `contacts.csv` CSV file.

### Option 1: Use Pandas to create the contacts DataFrame.

```python
import json

# Initialize an empty list to store contact dictionaries
contact_dicts = []

# Iterate through each row in the DataFrame
for index, row in contact_info_df.iterrows():
    # Extract JSON string from 'contact_info' column
    contact_info_json = row['contact_info']
    
    # Check if the JSON string is not empty
    if contact_info_json.strip():
        # Convert JSON string to dictionary
        try:
            contact_dict = json.loads(contact_info_json)
            contact_dicts.append(contact_dict)
        except json.JSONDecodeError as e:
            # Print the error message if JSON decoding fails
            print(f"Error decoding JSON at index {index}: {e}")
    else:
        # Print a message if the JSON string is empty
        print(f"Empty JSON string at index {index}")

# Convert the list of contact dictionaries to a DataFrame
contacts_df = pd.DataFrame(contact_dicts)

# Check the resulting DataFrame
print(contacts_df.head())
```

```python
# Create DataFrame from the list of dictionaries
contact_info_df = pd.DataFrame(contacts_df)

# Display the DataFrame
print(contact_info_df)
```

```python
# Create a contact_info DataFrame and add each list of values, i.e., each row 
# to the 'contact_id', 'name', 'email' columns.
contact_info = pd.DataFrame(contact_info_df, columns=['contact_id','name', 'email'])
contact_info.head()
# Print the contact_info DataFrame
print(contact_info)
```

```python
# Check the datatypes.
print(contact_info_df.dtypes)
```

```python
# Create a "first"name" and "l# Split the names into first name and last name
# Drop the contact_name column

# Split the names into first name and last name
contact_info[['first_name', 'last_name']] = contact_info['name'].str.split(' ', n=1, expand=True)

# Drop the original 'name' column
contact_info.drop(columns=['name'], inplace=True)

# Display the updated DataFrame
print(contact_info)




```

```python
# Reorder the columns
contact_info = contact_info[['contact_id', 'first_name', 'last_name', 'email']]

# Display the updated DataFrame
print(contact_info)
```

```python
# Check the datatypes one more time before exporting as CSV file.
print(contact_info.dtypes)
```

```python
# Export the DataFrame as a CSV file. 
contact_info.to_csv("Resources/contacts.csv", encoding='utf8', index=False)
```
